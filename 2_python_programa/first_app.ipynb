```python
a = "змінна з текстом"
b = 1
c = ["a", 1, 1.25, "Слово"]
d = {"a": "Слово", "b": 1}
e = ("a", )
f = {"ss", }

# Виведення значень змінних
print("Змінна 'a':", a)
print("Змінна 'b':", b)
print("Змінна 'c':", c)
print("Змінна 'd':", d)
print("Змінна 'e':", e)
print("Змінна 'f':", f)

print("Перша константа True:", True)
print("Друга константа None:", None)
```

```python
print("Абсолютне значення числа:", abs(-12.5))
print(f"Максимальне значення серед 5, 10 та 3: {max(5, 10, 3)}")
```

```python
letters = ["a", "b", "c"]
for i in range(len(letters)):
    print(f"На позиції {i} знаходиться буква {letters[i]}")

while letters:
    print(f"Залишилася буква: {letters.pop()}")
```

```python
A = True
print("Значить A=True" if A else "Значить A=False")

number = 42
result = "Парне" if number % 2 == 0 else "Непарне"
print(f"Число {number} є {result}")
```
