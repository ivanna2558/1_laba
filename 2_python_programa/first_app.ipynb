```python
# Задання змінних
a = "змінна з текстом"
b = 1  # числова Змінна
c = ["a", 1, 1.25, "Слово"]  # List
d = {"a": "Слово", "b": 1}  # Dict
e = ("a",)  # Tuple
f = {"ss", }  # Set

# Виведення значень змінних
print("Змінна 'a':", a)
print("Змінна 'b':", b)
print("Змінна 'c':", c)
print("Змінна 'd':", d)
print("Змінна 'e':", e)
print("Змінна 'f':", f)

# Виведення вбудованих констант
print("Пі значення (pi):", 3.141592653589793)
print("Імагінарна одиниця (j):", 1j)

# Використання вбудованих функцій
print("Абсолютне значення -12.5:", abs(-12.5))
print(f"Абсолютне значення 12.5 є рівним {abs(12.5)}")
```
